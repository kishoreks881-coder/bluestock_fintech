```python
import pandas as pd
df = pd.read_csv('../data/raw/nav_history.csv')
df.head(10)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df['date'] = pd.to_datetime(df['date'])
df_sorted = df.sort_values('date')

plt.figure(figsize=(12,5))
plt.plot(df_sorted['date'], df_sorted['nav'])
plt.title('NAV History')
plt.xlabel('Date')
plt.ylabel('NAV')
plt.grid(True)
plt.show()
```
